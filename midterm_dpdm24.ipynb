```python
# Load the data from the sheet "บ้านมือ2"
data_second_hand = house_data.parse('บ้านมือ2')

# Display basic information about the dataset
data_second_hand.info()

# Display the first few rows to understand the structure
data_second_hand.head()
```

```python
# Rounding the latitude and longitude to the nearest integer and then converting to int
data_second_hand['latitude'] = data_second_hand['latitude'].round().astype('Int64')
data_second_hand['longitude'] = data_second_hand['longitude'].round().astype('Int64')

# Check the data after conversion
data_second_hand[['latitude', 'longitude']].head()

```

```python
data_second_hand[['latitude', 'longitude']].info()
```

```python
# Select the relevant columns for analysis
columns_needed = [
    'renovate', 'number_of_rai', 'number_of_ngan', 'number_of_va', 'number_of_story',
    'number_of_bedroom', 'number_of_bathroom', 'direction', 'furniture', 'usable_area',
    'number_of_parking', 'sell_price', 'district', 'sub_district', 'zone'
]

# Filter the dataset with only the required columns
data_filtered = data_second_hand[columns_needed]

# Check missing values for each column
missing_values = data_filtered.isnull().sum()

# Display the missing values
missing_values
```

```python
# Handling missing values for numerical columns
data_filtered['number_of_story'].fillna(data_filtered['number_of_story'].median(), inplace=True)
data_filtered['usable_area'].fillna(data_filtered['usable_area'].median(), inplace=True)
data_filtered['number_of_parking'].fillna(data_filtered['number_of_parking'].median(), inplace=True)

# Handling missing values for categorical columns
data_filtered['renovate'].fillna(data_filtered['renovate'].mode()[0], inplace=True)
data_filtered['zone'].fillna(data_filtered['zone'].mode()[0], inplace=True)
data_filtered['sub_district'].fillna(data_filtered['sub_district'].mode()[0], inplace=True)

# Drop rows with missing 'sell_price' (only 1 missing value)
data_filtered.dropna(subset=['sell_price'], inplace=True)

# Verify missing values have been handled
missing_values_after = data_filtered.isnull().sum()

# Display the results
missing_values_after
```

```python
# เติมค่าขาดหายไปใน number_of_bedroom ด้วยค่ามัธยฐาน
data_filtered['number_of_bedroom'].fillna(data_filtered['number_of_bedroom'].median(), inplace=True)

# ตรวจสอบค่าที่ขาดหายไปอีกครั้ง
print(data_filtered.isnull().sum())
```

* ประเมินราคาซื้อขาย และ คะแนนทรัพย์ (ตั้งชื่อ 'estimator' และ 'score') จากไฟล์ "คำตอบนักประเมิน(ID)" - บ้าน โดยราคาประเมินกับคะแนนทรัพย์ อ้างอิงของนักประเมิน ID=01 คนเดียว (3 คะแนน)

```python
# prompt: ประเมินราคาซื้อขาย และ คะแนนทรัพย์ (ตั้งชื่อ 'estimator' และ 'score') จากไฟล์ "คำตอบนักประเมิน(ID)" - บ้าน โดยราคาประเมินกับคะแนนทรัพย์ อ้างอิงของนักประเมิน ID=01 คนเดียว

# Load the appraiser data
appraiser_data = pd.read_excel('/content/drive/MyDrive/Midterm_bsc_dpdm24_data/คำตอบนักประเมิน(ID) - บ้าน.xlsx')

# Filter data for appraiser ID 01
appraiser_01_data = appraiser_data[appraiser_data['ID'] == '01']

# Merge the appraiser's estimations with the house data
data_merged = pd.merge(data_filtered, appraiser_01_data, on='id', how='left')

# Rename columns
data_merged.rename(columns={'ราคาประเมิน': 'estimator', 'คะแนนทรัพย์': 'score'}, inplace=True)

# Handle missing values in 'estimator' and 'score' (if any) - You might choose imputation or deletion based on your analysis.
# Example: Impute with median
data_merged['estimator'].fillna(data_merged['estimator'].median(), inplace=True)
data_merged['score'].fillna(data_merged['score'].median(), inplace=True)


# Display the merged data
print(data_merged[['id', 'estimator', 'score']].head())
```

##2. จากตารางในข้อ 1 ตอบคำถามต่อไปนี้ (5 คะแนน)
zone ใดมีราคาซื้อขายบ้าน 'sell price' เฉลี่ยสูงที่สุด (1 คะแนน)
บ้านที่มีจำนวน 7-11 อยู่ในรัศมี 1 กิโลเมตรมากที่สุด ราคาเท่าไหร่ (2 คะแนน)
ตำบล (sub_district) ไหน ได้คะแนนทรัพย์ 'score' เฉลี่ยจากนักประเมินมากที่สุด (2 คะแนน)

```python
# prompt: zone ใดมีราคาซื้อขายบ้าน 'sell price' เฉลี่ยสูงที่สุด (1 คะแนน) บ้านที่มีจำนวน 7-11 อยู่ในรัศมี 1 กิโลเมตรมากที่สุด ราคาเท่าไหร่ (2 คะแนน) ตำบล (sub_district) ไหน ได้คะแนนทรัพย์ 'score' เฉลี่ยจากนักประเมินมากที่สุด

# zone ใดมีราคาซื้อขายบ้าน 'sell price' เฉลี่ยสูงที่สุด (1 คะแนน)
avg_sell_price_by_zone = data_merged.groupby('zone')['sell_price'].mean()
zone_with_highest_avg_price = avg_sell_price_by_zone.idxmax()
print(f"Zone with the highest average selling price: {zone_with_highest_avg_price}")

# บ้านที่มีจำนวน 7-11 อยู่ในรัศมี 1 กิโลเมตรมากที่สุด ราคาเท่าไหร่ (2 คะแนน)
house_with_most_711 = data_merged.loc[data_merged['d1000-7/11'].idxmax()]
price_of_house_with_most_711 = house_with_most_711['sell_price']
print(f"The house with the most 7-11s within 1km has a price of: {price_of_house_with_most_711}")


# ตำบล (sub_district) ไหน ได้คะแนนทรัพย์ 'score' เฉลี่ยจากนักประเมินมากที่สุด (2 คะแนน)
avg_score_by_subdistrict = data_merged.groupby('sub_district')['score'].mean()
subdistrict_with_highest_avg_score = avg_score_by_subdistrict.idxmax()
print(f"Subdistrict with the highest average appraiser score: {subdistrict_with_highest_avg_score}")
```
