```python
import pandas as pd

data = [[1, 2], [-2, 4], [0, 3], [-3, -2]]
df = pd.DataFrame(data, columns=['X1','X2'])
df
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is already defined from the previous code

plt.figure(figsize=(8, 6))
plt.scatter(df['X1'], df['X2'])
plt.plot([-3,1],[0,0], color='black', linewidth=2)  # Bold line for X-axis
plt.plot([0,0],[-2,4], color='black', linewidth=2)  # Bold line for Y-axis


plt.xlabel('X1', fontweight='bold')
plt.ylabel('X2', fontweight='bold')
plt.title('Scatter Plot of X1 vs X2')
plt.grid(True)
plt.show()
```

```python
# prompt: calculate distance matrix of the df

import pandas as pd
from scipy.spatial.distance import pdist, squareform

# Assuming your DataFrame 'df' has numerical columns for which you want to calculate the distance matrix
#Example DataFrame (replace with your actual DataFrame)
data = [[1, 2], [-2, 4], [0, 3], [-3, -2]]
df = pd.DataFrame(data, columns=['X1','X2'])


# Convert the condensed distance matrix to a square distance matrix
distance_matrix = squareform(pdist(df[['X1','X2']]))
distance_matrix

```

```python
# prompt: calculate distance matrix of the df use L1 norm


# Calculate the distance matrix using the L1 norm (Manhattan distance)

distance_matrix_l1 = squareform(pdist(df[['X1', 'X2']], metric='cityblock'))  # Use 'cityblock' for L1 norm
distance_matrix_l1
```

#Normalization (Z-core & Min-Max normalization)

#Z-score

```python
# prompt: plot histogram of df2 X2

import matplotlib.pyplot as plt

# Assuming df2 is already defined from the previous code
plt.hist(df2['X2'], bins=10)  # Adjust the number of bins as needed
plt.xlabel('X2')
plt.ylabel('Frequency')
plt.title('Histogram of df2 X2')
plt.show()
```

```python
# prompt: calculate Z-score from df2

import pandas as pd
from scipy.stats import zscore

# Assuming df2 is already defined from the previous code
# Calculate Z-scores for all columns in df2
df2_zscore = df2.apply(zscore)

# Print or use the DataFrame with Z-scores
df2_zscore
```

```python
# prompt: do min-max normalization for the df2 for each column separately

# Assuming df2 is already defined from the previous code
# Min-Max normalization for each column separately

for column in df2.columns:
    min_val = df2[column].min()
    max_val = df2[column].max()
    df2[column + '_normalized'] = (df2[column] - min_val) / (max_val - min_val)

df2
```
